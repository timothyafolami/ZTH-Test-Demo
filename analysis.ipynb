```python
from rouge import Rouge

# Example lists of generated sentences and reference sentences
generated_sentences = ["This is a sample summary 1.", "This is a sample summary 2."]
reference_sentences = ["This is a sample summary 1.", "This is a reference summary 2."]

# Initialize Rouge
rouge = Rouge()

# Initialize variables to store aggregated Rouge scores
total_rouge_1_r = 0
total_rouge_2_r = 0
total_rouge_l_r = 0

# Compute Rouge scores for each pair of sentences
for gen_sentence, ref_sentence in zip(generated_sentences, reference_sentences):
    scores = rouge.get_scores(gen_sentence, ref_sentence)[0]
    total_rouge_1_r += scores['rouge-1']['r']


# Calculate average Rouge scores
num_sentences = len(generated_sentences)
avg_rouge_1_r = total_rouge_1_r / num_sentences

# Normalize Rouge scores to a scale of 0 to 100%
normalized_rouge_1_r = avg_rouge_1_r * 100


print("Normalized Rouge-1 Recall:", normalized_rouge_1_r)


# putting this in a function
def calculate_rouge_scores(generated_sentences, reference_sentences):
    # Initialize Rouge
    rouge = Rouge()

    # Initialize variables to store aggregated Rouge scores
    total_rouge_1_r = 0
    total_rouge_2_r = 0
    total_rouge_l_r = 0

    # Compute Rouge scores for each pair of sentences
    for gen_sentence, ref_sentence in zip(generated_sentences, reference_sentences):
        scores = rouge.get_scores(gen_sentence, ref_sentence)[0]
        total_rouge_1_r += scores['rouge-1']['r']

    # Calculate average Rouge scores
    num_sentences = len(generated_sentences)
    avg_rouge_1_r = total_rouge_1_r / num_sentences

    # Normalize Rouge scores to a scale of 0 to 100%
    normalized_rouge_1_r = avg_rouge_1_r * 100

    return normalized_rouge_1_r
```
